## Data Summary

The dataset encompasses vehicles from different eras, providing a comprehensive overview of the evolution of electric vehicles (EVs) throughout time. It features essential information such as a unique Vehicle Identification Number (VIN), along with registration details like county, city, and postal code. Additionally, it records the model year, make, and model of each vehicle. By differentiating between Battery Electric Vehicles (BEVs) and Plug-in Hybrid Electric Vehicles (PHEVs), the dataset showcases the increasing variety of EV types. It also includes information on Clean Alternative Fuel Vehicle (CAFV) eligibility, electric range, and base Manufacturer's Suggested Retail Price (MSRP), offering insights into the technological advancements and cost-effectiveness of electric vehicles. Furthermore, the dataset incorporates legislative district information, unique Department of Licensing (DOL) vehicle IDs, and vehicle locations, enhancing its value for analyzing the trends and progression in EV adoption over time.

## Data Description

**VIN**: The Vehicle Identification Number is a unique alphanumeric code assigned to each vehicle for identification purposes. This column represents the first 10 characters of the VIN.

**County**: The county where the vehicle is registered in Washington State.

**City**: The city where the vehicle is registered in Washington State.

**State**: The state where the vehicle is registered (Washington in this case).

**Postal Code**: The postal code of the registration address for the vehicle.

**Model Year**: The year in which the vehicle was manufactured.

**Make**: The manufacturer or brand of the vehicle.

**Model**: The specific model or name of the vehicle.

**Electric Vehicle Type**: Indicates whether the vehicle is a Battery Electric Vehicle (BEV) or a Plug-in Hybrid Electric Vehicle (PHEV).

**Clean Alternative Fuel Vehicle (CAFV) Eligibility**: Indicates if the vehicle meets the eligibility criteria for Clean Alternative Fuel Vehicle incentives or benefits.

**Electric Range**: The distance the vehicle can travel on electric power alone, typically measured in miles.

**Base MSRP**: The Manufacturer's Suggested Retail Price, which is the starting price set by the vehicle manufacturer.

**Legislative District**: The legislative district associated with the vehicle's registered address.

**DOL Vehicle ID**: A unique identifier assigned by the Washington State Department of Licensing (DOL) for each registered vehicle.

**Vehicle Location**: The precise location of the vehicle, which could be the address or coordinates.

**Electric Utility**: The name of the electric utility company associated with the vehicle, if applicable.

**2020 Census Tract**:The census tract associated with the vehicle's registered address, based on the 2020 Census data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define a custom color palette for the pie chart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

# Create subplots in a 4x2 layout to minimize rows
fig, axs = plt.subplots(4, 2, figsize=(20, 30))

# 1. Pie Chart for Electric Vehicle Types with count labels
vehicle_type_counts = df['Electric Vehicle Type'].value_counts()
axs[0, 0].pie(vehicle_type_counts, labels=[f'{label}: {count}' for label, count in zip(vehicle_type_counts.index, vehicle_type_counts)], 
              autopct='%1.1f%%', startangle=90, colors=colors)
axs[0, 0].set_title('Distribution of Electric Vehicle Types', fontweight='bold')

# 2. Box Plot for Base MSRP by Electric Vehicle Type
sns.boxplot(x='Electric Vehicle Type', y='Base MSRP', data=df, ax=axs[0, 1], palette='pastel')
axs[0, 1].set_title('Base MSRP by Electric Vehicle Type', fontweight='bold')

# 3. Histogram for Electric Range
sns.histplot(df['Electric Range'], bins=20, kde=True, color='skyblue', ax=axs[1, 0])
axs[1, 0].set_title('Electric Range Distribution', fontweight='bold')
axs[1, 0].set_xlabel('Electric Range')
axs[1, 0].set_ylabel('Frequency')

# 4. Count Plot for Electric Vehicle Type by Model Year
sns.countplot(data=df, x='Model Year', hue='Electric Vehicle Type', palette='viridis', ax=axs[1, 1])
axs[1, 1].set_title('Count of Electric Vehicle Types by Model Year', fontweight='bold')
axs[1, 1].set_xlabel('Model Year')
axs[1, 1].set_ylabel('Count')
axs[1, 1].tick_params(axis='x', rotation=90)

# Annotate counts on bars with extra space and bold font
for p in axs[1, 1].patches:
    axs[1, 1].annotate(f'{int(p.get_height())}', 
                       (p.get_x() + p.get_width() / 2., p.get_height() + 1), 
                       ha='center', va='bottom', fontsize=12, fontweight='bold', rotation=90)

# 5. Scatter Plot for Electric Range vs Base MSRP
sns.scatterplot(x='Electric Range', y='Base MSRP', hue='Electric Vehicle Type', data=df, ax=axs[2, 0], palette='Set1')
axs[2, 0].set_title('Electric Range vs Base MSRP', fontweight='bold')

# 6. Boxplot for Model Year vs Base MSRP
sns.boxplot(x='Model Year', y='Base MSRP', data=df, ax=axs[2, 1], palette='pastel')
axs[2, 1].set_title('Model Year vs Base MSRP', fontweight='bold')
axs[2, 1].tick_params(axis='x', rotation=45)

# 7. Correlation Heatmap
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, ax=axs[3, 0])
axs[3, 0].set_title('Correlation Heatmap', fontweight='bold')

# Remove empty subplot (bottom right)
fig.delaxes(axs[3, 1])

# Adjust layout to ensure the plots don't overlap
plt.tight_layout()
plt.show()
```

### 7. **Correlation Heatmap - Numeric Features**:

# Creating a choropleth map using Plotly Express

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# Sample DataFrame (assuming df has 'State', 'Model Year', and 'Electric Vehicle Type' columns)
# Replace this with your actual data
# df = pd.read_csv('your_data.csv')  

# Grouping data by state and model year to get the count of EVs
ev_count_by_state_year = df.groupby(['State', 'Model Year']).size().reset_index(name='EV Count')

# Create a base choropleth map
fig = px.choropleth(
    ev_count_by_state_year,
    locations='State',
    locationmode='USA-states',
    color='EV Count',
    scope="usa",
    color_continuous_scale="YlGnBu",
    labels={'EV Count': 'Number of EVs'},
)

# Updating layout for title and aesthetics
fig.update_layout(
    title={
        'text': "Number of EV Vehicles by State",
        'font': {'size': 20, 'color': 'black', 'family': 'Arial', 'weight': 'bold'},
        'x': 0.5,  # Centering the title
        'xanchor': 'center',
        'yanchor': 'top'
    },
    margin=dict(l=10, r=10, t=40, b=10)  # Adjust margins to reduce white space
)

# Making the outlines of the states bold
fig.update_geos(
    showcountries=True,
    coastlinecolor="Black",
    countrycolor="Black",
    landcolor="lightgray",
    subunitcolor="black",  # State borders
    subunitwidth=2,  # Width of the state borders
)

# Create a list of unique years for the dropdown
years = sorted(ev_count_by_state_year['Model Year'].unique())

# Adding traces for each year for the dropdown
for year in years:
    filtered_data = ev_count_by_state_year[ev_count_by_state_year['Model Year'] == year]
    fig.add_trace(
        go.Choropleth(
            locations=filtered_data['State'],
            locationmode='USA-states',
            z=filtered_data['EV Count'],
            colorscale='YlGnBu',
            showscale=False,
            visible=(year == years[0]),  # Only the first year is visible initially
            name=str(year)
        )
    )

# Adding a trace for "All Years"
total_ev_counts = ev_count_by_state_year.groupby('State')['EV Count'].sum().reset_index()
fig.add_trace(
    go.Choropleth(
        locations=total_ev_counts['State'],
        locationmode='USA-states',
        z=total_ev_counts['EV Count'],
        colorscale='YlGnBu',
        showscale=False,
        visible=False,  # Initially not visible
        name="All Years"
    )
)

# Create dropdown buttons
dropdown_buttons = []
# Button for "All Years"
dropdown_buttons.append({
    'label': 'All Years',
    'method': 'update',
    'args': [
        {'visible': [True] + [False] * len(years)},  # Show only the All Years trace
        {'title': "Total Number of EV Vehicles by State (All Years)"}
    ]
})

# Buttons for each year
for year in years:
    button = {
        'label': str(year),
        'method': 'update',
        'args': [
            {'visible': [year == y for y in years] + [False]},  # Update visibility
            {'title': f"Number of EV Vehicles by State in {year}"}
        ]
    }
    dropdown_buttons.append(button)

# Adding the dropdown menu to the layout
fig.update_layout(
    updatemenus=[{
        'buttons': dropdown_buttons,
        'direction': 'down',
        'showactive': True,
        'x': 0.08,  # Position of the dropdown on the x-axis
        'y': 0.95,  # Position of the dropdown on the y-axis
    }]
)

# Display the figure
fig.show()
```
